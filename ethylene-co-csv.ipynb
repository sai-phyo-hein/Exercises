```python
file = open(dirname + "/ethylene_CO.txt")
CO_lines = file.readlines()
CO_lines.pop(0)
```

```python
data_list = []
def formater(lines):
  for line in lines:
    txt = line.split(" ")
    dummy = []
    for t in txt:
      if t != '':
        dummy.append(t)
        dummy[-1] = dummy[-1].replace("\n", "")
    data_list.append(dummy)
```

```python
formater(CO_lines)
```

```python
data= pd.DataFrame(data_list, columns =  ['Time(sec)',
 'CO con: (ppm)',
 'Ethylene con: (ppm',
 'TGS2602-1',
 'TGS2602-2',
 'TGS2600-1',
 'TGS2600-2',
 'TGS2610-1',
 'TGS2610-2',
 'TGS2620-1',
 'TGS2620-2',
 'TGS2602-3',
 'TGS2602-4',
 'TGS2600-3',
 'TGS2600-4',
 'TGS2610-3',
 'TGS2610-4',
 'TGS2620-3',
 'TGS2620-4',""])
```

```python
data = data.drop([""],axis = 1)
data.head()
```
